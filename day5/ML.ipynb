```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("C:\sam\day5\pass.csv")
print(df.head())

plt.figure(figsize=(8,6))
plt.scatter(df["hours"],df["pass"], c=df["pass"],cmap = "bwr",edgecolors='k')
plt.xlabel("hours")
plt.ylabel("pass or fail")
plt.title("simple Dataset")
plt.show()
```

```python
x = df[["hours"]]
y = df[["pass"]]

model = LogisticRegression()
model.fit(x,y)
print("model trained successfullly")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["mark"], df["pass"], c= df ["pass"], cmap="bwr",label = "data points")

x_test_range = np.linspace(0,10,100).reshape(-1,1)
y_prob_range = model.predict_proba(x_test_range)[:,1]
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['hours'], df['pass'], c=df['pass'], cmap='bwr', label='Data Points')
X_test_range = np.linspace(0, 10, 100).reshape(-1, 1)
y_prob_range = model.predict_proba(X_test_range)[:, 1]
plt.plot(X_test_range, y_prob_range, color='green', linewidth=3, label='Logistic Curve')
plt.axhline(y=0.5, color='gray', linestyle='--', label='Decision Boundary (0.5)')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression Sigmoid Curve')
plt.legend()
plt.show()
```

```python
try:
    user_hours = float(input("Enter study hours (0-10): "))
    X_user = np.array([[user_hours]])
    prediction = model.predict(X_user)[0]
    probability = model.predict_proba(X_user)[0][1]
    if prediction == 1:
        print("\nResult: PASS")
        print(f"Confidence: {probability*100:.2f}%")
    else:
        print("\nResult: FAIL X")
        print(f"Confidence: {(1-probability)*100:.2f}%")

except ValueError:
    print("Invalid input. Please enter a numerical value for hours.")
```
